### Reservations Canceled or not of Different Hotels

```python
#Check which Hotels have high & low cancelation rates

plt.figure(figsize=(7,5))
ax1 = sns.countplot(x='hotel', hue='is_canceled', data=dataset, palette='PuBuGn')
legend_labels, _ = ax1.get_legend_handles_labels()

# Use Matplotlib's legend function
ax1.legend(legend_labels, ['Not Canceled', 'Canceled'], bbox_to_anchor=(1, 1))

plt.xlabel('HOTEL')
plt.ylabel('Number of Reservations')
plt.show()
```

### Months Having Reservation and then Cancelation / Not Cancelation

```python
#Now look at months where there are reservtions and where there are cancelations
#we don't have any month column
#we will make month column using the date column
dataset['Month']=dataset['reservation_status_date'].dt.month


plt.figure(figsize=(12,5))
ax1 = sns.countplot(x='Month', hue='is_canceled', data=dataset, palette='twilight')
legend_labels, _ = ax1.get_legend_handles_labels()

# Use Matplotlib's legend function
ax1.legend(legend_labels, ['Not Canceled', 'Canceled'], bbox_to_anchor=(1, 1))
plt.title('Reservation Status per Month')
plt.xlabel('Months')
plt.ylabel('Number of Reservations')
plt.show()
```

### Cancelation rate Month-wise with respect to 'adr'

```python
# This is only for cancellation rate
# beacuse we want to check whether price is affecting cancelation or not

plt.figure(figsize=(7,5))
plt.title('ADR per month')
sns.barplot('Month','adr',data=dataset[dataset['is_canceled']==1].groupby('Month')[['adr']].sum().reset_index())
#dataset['is_canceled']==1  means only canceled values
plt.show()
```

Our hypo proves that if price is high than cancelation is also high
They do reservation(as shown in previous graph of Reservation status) and maybe at last moment they cancel it

### Cancelation Rate - country wise

```python
#only see top 10 countries , as there are 177 countries in data

cancelled_data=dataset[dataset['is_canceled']==1]
top_10_countries=cancelled_data['country'].value_counts()[:10]  #sort in decending
#value_counts means how many times the value is coming in that column

plt.figure(figsize=(8,8))

plt.title('Top 10 Countries with Reservation Cancelled')

plt.pie(top_10_countries,autopct='%.2f',labels=top_10_countries.index)
#autopct='%.2f' values upto 2 decimals
plt.show()
```

### Cancelation And Not Cancelation mean wrt adr

```python
# cancelled_data=dataset[dataset['is_canceled']==1]
not_cancelled_data=dataset[dataset['is_canceled']==0]
```

```python

cancelled_dataset_adr=cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_dataset_adr.reset_index(inplace=True)
cancelled_dataset_adr.sort_values('reservation_status_date',inplace=True)

not_cancelled_dataset_adr=not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_dataset_adr.reset_index(inplace=True)
not_cancelled_dataset_adr.sort_values('reservation_status_date',inplace=True)
```

```python
plt.figure(figsize=(18,7))
plt.plot(cancelled_dataset_adr['reservation_status_date'],cancelled_dataset_adr['adr'],label='Cancelled')
plt.plot(not_cancelled_dataset_adr['reservation_status_date'],not_cancelled_dataset_adr['adr'],label='Not Cancelled')
plt.title('Average Daily Rate')
plt.legend()
plt.show()
```

**cancelled reservation mean is high, or not canceled reservation mean is high**

```python
plt.figure(figsize=(18,7))
plt.scatter(cancelled_dataset_adr['reservation_status_date'],cancelled_dataset_adr['adr'],label='Cancelled')
plt.scatter(not_cancelled_dataset_adr['reservation_status_date'],not_cancelled_dataset_adr['adr'],label='Not Cancelled')
plt.title('Average Daily Rate')
plt.legend()
plt.show()
```

**We have more data in middle , less at start and end**

**There is in-consistency in dataset**

**We will remove it by taking specific range***

**Take only 2016-01 to 2017-09 data**

### Taking Data from specific Range

```python
cancelled_dataset_adr=cancelled_dataset_adr[(cancelled_dataset_adr['reservation_status_date']>'2016-01') 
                                            & (cancelled_dataset_adr['reservation_status_date']<'2017-09')]


not_cancelled_dataset_adr=not_cancelled_dataset_adr[(not_cancelled_dataset_adr['reservation_status_date']>'2016-01') 
                                                    & (not_cancelled_dataset_adr['reservation_status_date']<'2017-09')]
```

```python
plt.figure(figsize=(18,7))
plt.plot(cancelled_dataset_adr['reservation_status_date'],cancelled_dataset_adr['adr'],label='Cancelled')
plt.plot(not_cancelled_dataset_adr['reservation_status_date'],not_cancelled_dataset_adr['adr'],label='Not Cancelled')
plt.title('Average Daily Rate',fontsize=18)
plt.legend(fontsize=18)
plt.show()
```

**Avg Daily rate of canceled is more than AVg Daily rate of not cancelled**

**Proof --> Cancellation depends on Price**

**Spikes are nothing but weekend or holidays**
